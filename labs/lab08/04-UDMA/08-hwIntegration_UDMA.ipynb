```python
# Length of the input signal 
TRACE_LEN = 201

# Input registers
CB_INFERENCE_REG = 0 # Inference result
```

```python
#IP_ADDRESS = 'XXX.XXX.XXX.XXX' # Set your ZedBoard IP Address here
IP_ADDRESS = '192.168.1.10' 
IP_PORT = 7 
```

```python
board = udma.UDMA_CLASS(IP_ADDRESS, IP_PORT) 
```

```python
connectionStatus = 0
while(connectionStatus == 0):
    connectionStatus = board.connect()
    # sleep(1)
```

```python
board.log(0)
```

```python
# Function definition: used to clear out FIFO contents prior to read any value
def clearInFifo(udmaInstance):
    udmaInstance.write_reg(17, 1)
    udmaInstance.write_reg(17, 0)
```

#### Se√±al a ser enviada a la FPGA

```python
signal = []
```

#### Interaction with ComBlock to send the signal and receive the inference result

```python
# Clear input FIFO contents to get the latest data from the FIR filter output
clearInFifo(board)
```

```python
# Writing to ComBlock's FIFO the signal to be used as input for the inference IP core
# signal should be array-type

board.write_fifo(length=TRACE_LEN, Txdata=signal)[0] # <--- This is the signal variable that will be sent
```

#### Resultado de la inferencia

```python
# Read inference result
board.read_reg(reg=CB_INFERENCE_REG)[1][0]
```
